```python
import pandas as pd
import numpy as np
```

```python
info={'Company':['Ford','Ford','VW','BMW','Cooper','Cooper'],
'Stars':[1,2,np.nan,2,1,1],
'Weight':[2,4,2,2,3,None],
'Origin': ['China','Mexico','Mexico',None, 'China',pd.NaT],
'Length':[40,50,30,np.nan,45,None]
}
```

```python
df=pd.DataFrame(data=info)
df.head()
```

```python
df.isnull().any()
```

```python
ndf=df.copy()
ndf.dropna(axis=1)
```

```python
sm=df.Stars.mean()
```

```python
df['Stars'].fillna(value=sm,
                   inplace=True)
df
```

```python
dx=df.Weight.mean()
df['Weight'].fillna(value=dx,
                   inplace=True)
df
```

```python
ov=df.Origin.mode()[1]
ov
```

```python
df.Origin.mode()
```

```python
df['Origin'].fillna(value=ov,inplace =True)
df
```

```python
Le=df.Length.mean()
df['Length'].fillna(value=Le,
                   inplace=True)
df
```

```python
df.to_csv('test.csv')
```

```python
##What is the avergae stars of cars made in mexico?
```

```python
pd.pivot_table(df,
         index='Origin',
          values='Stars',
          aggfunc=['sum','mean','median']
)
```

```python
my_pivot=pd.pivot_table(df,
         index=['Origin','Company'],
          values='Stars',
          aggfunc=['mean']
         # aggfunc=['sum','mean','median']
          #columns='Company'
)
my_pivot
```
