```python
path='/content/drive/MyDrive/Canada.xlsx'


```

```python
import pandas as pd
import numpy as np
df=pd.read_excel(path,
                sheet_name='Canada by Citizenship',
                 skiprows=range(20),
                 skipfooter=2
                 )
df.head()
df.tail()
```

```python
no_favs=['Type',	'Coverage',	'AREA',	'REG',	'DEV',	'DevName']
df.drop(no_favs, axis=1, inplace=True)
df.head(2)
```

```python
new_names={'OdName':'Country','AreaName':'Continent', 'RegName':'Region'}
df.rename(columns=new_names,inplace=True)
df.head(2)
```

```python
df['Total']=df.iloc[:,3:].sum(axis=1)
df.head(2)
```

```python
df.iloc[80,:]
df.set_index('Country',inplace=True)
df.head(2)
```

```python
df.index.name=None
df.head(2)
```

```python
afgh_total=(df.loc['Afghanistan','Total'].sum(axis=0))
afgh_total
```
